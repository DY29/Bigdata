```python
import pandas as pd
import json
import warnings

warnings.simplefilter(action = "ignore", category = FutureWarning)
```

```python
geo_path = './data/seoul_data/skorea_municipalities_geo_simple.json'
geo_str  = json.load(open(geo_path, encoding='utf-8'))
```

```python
df = pd.read_csv('./data/seoul_data/seoul_wifi_zone.csv')
df.head()
```

```python
df.columns = (['구', '유형', '위치명', '위도', '경도', '설치회사'])
df.head()
```

```python
freq_gu = df['구'].value_counts()
freq_gu
```

```python
freq_gu.describe()
```

```python
map = folium.Map(location=[37.5102, 126.982], zoom_start=11)
map
```

```python
map = folium.Map(location = [37.5102, 126.982], tiles= 'Stamen Toner', zoom_start=11)
map
```

```python
map = folium.Map(location=[37.5502, 126.982], tiles='Stamen Toner', zoom_start=11)

map.choropleth(geo_data=geo_str,
               data=freq_gu,
               columns=['gu', 'counts'],
               fill_color='YlGnBu', #PuRd, YlGnBu
               key_on='feature.id')
map
```

```python
population['고령비율'] = population['65세이상고령자'] / population['인구수'] * 100
population.head()
```

```python
guDat = pd.DataFrame({'gu':population['구'], 'counts':population['고령비율']})
map = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Stamen Toner')

map.choropleth(geo_data=geo_str,
              data=guDat,
              columns=['gu', 'counts'],
              fill_color='PuRd', #PuRd, YlGnBu
              key_on='feature.id')
map
```
